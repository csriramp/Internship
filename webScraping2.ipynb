```python
#Question 1 Scraping Data from naukri for data analyst 
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url='https://www.naukrigulf.com/?showQsb=1'
driver.maximize_window()
driver.get(url)
driver.implicitly_wait(10)

driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/form/div[1]/input').send_keys('Data Analyst')
driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/form/div[2]/input').send_keys('Dubai')
driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/form/input').click()
driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/main/div/section[2]/div[1]/form/div[3]/div[2]/ul/li[1]/label/span[2]').click()
titles=driver.find_elements(By.XPATH,'//h2[@class="designation-title"]')
location=driver.find_elements(By.XPATH,'//li[contains(@class,"info-loc")]')
company=driver.find_elements(By.XPATH,'//*[contains(@class,"info-org")]')
experience_tag=driver.find_elements(By.XPATH,'//li[contains(@class,"info-exp")]')

job_titles=[]
job_location=[]
company_name=[]
experience=[]

for i,j,k,l in zip(titles,location,company,experience_tag):
    job_titles.append(i.text)
    job_location.append(j.text)
    company_name.append(k.text)
    experience.append(l.text)

naukri_da_df=pd.DataFrame({'Job Title':job_titles[0:10],'Job Location':job_location[0:10],'Comany':company_name[0:10],'Relevant experience required':experience[0:10]})
naukri_da_df
```

```python
#Question 4 Sunglasses from flipkart
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url='https://www.flipkart.com/'
driver.maximize_window()
driver.get(url)
driver.implicitly_wait(10)

#try catch for closing the registration popup
try:
    driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button').click()
except:
    pass

brand=[]
prod_desc=[]
price=[]

#Entering sunglasses and click 
driver.find_element(By.XPATH,'//input[@name="q"]').send_keys('sunglasses')
driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]').click()

#appending the branch , description and price details of page1 to the list
p1_brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
p1_desc=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
p1_price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')

for i,j,k in zip(p1_brand,p1_desc,p1_price):
    brand.append(i.text)
    prod_desc.append(j.text)
    price.append(k.text)
    
#clicking on next and appending the details of page2 to list
driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]').click()
time.sleep(1)
p2_brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
p2_desc=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
p2_price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')

for l,m,n in zip(p2_brand,p2_desc,p2_price):
    brand.append(l.text)
    prod_desc.append(m.text)
    price.append(n.text)

#clicking on next and appending the details of page3 to list
driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[12]').click()
time.sleep(1)
p3_brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
p3_desc=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
p3_price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')

for o,p,q in zip(p3_brand,p3_desc,p3_price):
    brand.append(o.text)
    prod_desc.append(p.text)
    price.append(q.text)
  
  

#Creating a data frame with top 100 search results

df_fk_sunglasses=pd.DataFrame({'Brand Name':brand[0:100],'Product Description':prod_desc[0:100],'Product Price':price[0:100]})

df_fk_sunglasses
```

```python
#Question 5 Iphone11 reviews
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url='https://www.flipkart.com/'
driver.maximize_window()
driver.get(url)
driver.implicitly_wait(10)
driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button').click()
driver.find_element(By.CLASS_NAME,"_3704LK").send_keys("iphone 11")
driver.find_element(By.CLASS_NAME,"L0Z3Pu").click()
driver.get('https://www.flipkart.com/apple-iphone-11-green-64-gb/product-reviews/itm386a85013a3d1?pid=MOBFWQ6BHPMG29QM&lid=LSTMOBFWQ6BHPMG29QMKXJSSF&marketplace=FLIPKART')

rating=[]
review_sum=[]
review_full=[]


def review_scrap():
    
    rate=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rate:
        rating.append(i.text)
        
    review=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in review:
        review_sum.append(i.text)
        
    full=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in full:
        review_full.append(i.text)

#scraping page numbers        
page_no=driver.find_elements(By.CLASS_NAME,"ge-49M")

#looping through each page for execution of scarping function and scrape 100 reviews
for j in page_no:
    review_scrap()

df_fk_review=pd.DataFrame({ 'Rating':rating,
                            'Review Summary':review_sum,
                            'Review Full':review_full})
df_fk_review
```

```python
#Question 8 Scraping from Amazon for i7 laptops
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.amazon.in'
driver.get(url)
driver.maximize_window()
driver.implicitly_wait(10)

description=[]
prod_price=[]
prod_rating=[]

driver.find_element(By.XPATH,'//div[@class="nav-search-field "]/input').send_keys('laptop') #providing search input
time.sleep(1)
#driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input').click()
driver.find_element(By.XPATH,'//input[@value="Go"]').click()

time.sleep(1)
driver.find_element(By.XPATH,"//span[contains(text(),'Intel Core i7')]").click() #selecting i7 checkbox
time.sleep(3)
desc=driver.find_elements(By.XPATH,'//span[contains(@class,"a-size-medium a-color-base a-text-normal")]') #Description scraping
price=driver.find_elements(By.XPATH,'//span[contains(@class,"price-whole")]') # Price Scraping
rating=driver.find_elements(By.XPATH,"//div[@class='a-row a-size-small']/span") #ratinig scraping
#rating=driver.find_elements(By.XPATH,'//span[@class="a-declarative"]//i[contains(@class,"aok-align-bottom")]//span[contains(@class,"a-icon-alt")]')
#rating=driver.find_elements(By.XPATH,'//span[@class="a-icon-alt"]')
#rating=driver.find_elements(By.XPATH,'//i[contains(@class,"aok-align-bottom")]/span[contains(@class,"a-icon-alt")]')
time.sleep(3)


for i,j,k in zip(desc,price,rating):
    description.append(i.text)
    prod_price.append(j.text)
    prod_rating.append(k.get_attribute("aria-label"))

    
prod_rating= [prod_rating[i] for i in range(len(prod_rating)) if i%2==0] #segregation of rating and count into only ratings

df_amazon=pd.DataFrame({'Description of product':description[0:10],
                        'Price of Product': prod_price[0:10],
                        'Rating of Product': prod_rating[0:10]})

df_amazon
```

```python
#Question 9 Scraping Data from Ambition box Noida

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
company_name=[]
company_rating=[]
days_posted=[]
url="https://www.ambitionbox.com/jobs"
driver.get(url)
driver.maximize_window()

#Entering data science in search
driver.find_element(By.NAME,"ab_jobsSearch").send_keys('Data Scientist')
#Clicking the search
driver.find_element(By.CSS_SELECTOR,"div[class='typeahead-wrapperjobsv3'] span[class='ctas-btn-medium']").click()
time.sleep(3)
#Selecting the location dropdown button
location=driver.execute_script("return document.getElementsByTagName('i')[16];")
driver.execute_script("arguments[0].click();",location)
time.sleep(1)
#Entering Noida in the text box
driver.find_element(By.XPATH,"//input[@placeholder='Search locations']").send_keys("Noida")
time.sleep(1)
#clicking Noida
driver.find_element(By.XPATH,"//label[@for='location_Noida']").click()
time.sleep(10)

# Scraping company name
name=driver.execute_script("return document.getElementsByClassName('company body-medium');")
rating=driver.execute_script("return document.getElementsByClassName('body-small');")
days=driver.find_elements(By.XPATH,"//span[@class='body-small-l'][1]")


    
for i,j,k in zip(name,rating,days):
    company_name.append(i.text)
    company_rating.append(j.text)
    days_posted.append(k.text)

df_ambition_box_9=pd.DataFrame({'Name of Company':company_name,'Rating':company_rating,'Posted':days_posted})

df_ambition_box_9
```

```python
#Question 10 Ambition box
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url='https://www.ambitionbox.com/'
driver.get(url)
driver.maximize_window()



#hover over salaries and select browse salaries
salary=driver.find_element(By.LINK_TEXT,"Salaries")
#creating hovering object 
hover=ActionChains(driver)
hover.move_to_element(salary).perform()
time.sleep(1)
#Selecting the browse salary option
driver.find_element(By.XPATH,"//a[@title='Browse salaries']").click()
time.sleep(1)
job_prof=driver.find_element(By.ID,"jobProfileSearchbox")
job_prof.send_keys("Data Scientist")

clk=driver.find_elements(By.XPATH,"//div[@class='suggestion_wrap tt-suggestion tt-selectable']")
for i in clk:
    if i.text == 'Data Scientist':   
        i.click() # click on “Data Scientist”
        break 

#scrape Company name, Number of salaries, Average salary, Min salary, Max Salary, experience required.

company_name = []
salary_record = []
average_salary = []
min_salary = []
max_salary = []
exp = []

#Scraing all tags

company=driver.find_elements(By.XPATH,"//div[@class='company-info']")
for i in company:
    company_name.append(i.text.split("\n")[0])                              #company name
    salary_record.append(i.text.replace("based on", " ").replace("salaries","").split("\n")[1:2]) #Total Salary Record
    exp.append(i.text.replace("yrs exp","").replace("yr exp","").split("\n")[-1])                # Experience required
    
avg_salary=driver.find_elements(By.XPATH,"//div[@class='average-indicator-wrapper']")
for i in avg_salary:
    average_salary.append(i.text.replace("₹","").replace("L",""))            # average salary

salary=driver.find_elements(By.XPATH,"//div[@class='salary-values']")
for i in salary:
    min_salary.append(i.text.replace("₹","").replace("L","").split("\n")[0])  # minimum salary
    max_salary.append(i.text.replace("₹","").replace("L","").split("\n")[1])  # maximum salary
    
# creating Dataframe 

df_ambition_box_10=pd.DataFrame({})
df_ambition_box_10["Company_Name"] = company_name[:10]
df_ambition_box_10["Total_Salary_Record"] = salary_record[:10]
df_ambition_box_10["Experience_Required"] = exp[:10]
df_ambition_box_10["Average_Salary_L"] = average_salary[:10]
df_ambition_box_10["Minimum_Salary_L"] = min_salary
df_ambition_box_10["Maximum_Salary_L"] = max_salary


df_ambition_box_10
```
